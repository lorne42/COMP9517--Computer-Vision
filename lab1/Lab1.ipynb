```python
#Task1
import cv2
import numpy as np


image = cv2.imread('./Task1.jpg')


gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


inverted_image = 255 - gray_image

gamma = 5  
gamma_corrected_image = np.power(inverted_image / 255.0, gamma) * 255.0
gamma_corrected_image = np.uint8(gamma_corrected_image)


cv2.namedWindow('Original Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Original Image', 900, 600)



cv2.namedWindow('Gamma Corrected Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Gamma Corrected Image', 900, 600)

cv2.imshow('Original Image', gray_image)
cv2.imshow('Gamma Corrected Image', gamma_corrected_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Task2
"""
I choose to use the median filter.
Using a median filter instead of a Gaussian filter is often preferred for several reasons:
Better removal of salt-and-pepper noise: 
Median filters excel at removing salt-and-pepper noise (random black and white pixels) compared to Gaussian filters.
They achieve this by replacing each pixel's value with the median value of its neighborhood, 
effectively eliminating extreme values without blurring edges.
"""
import cv2


image = cv2.imread('./Task2.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

median_filtered_image = cv2.medianBlur(gray_image, 5)  

cv2.namedWindow('Original Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Original Image', 600, 700)

cv2.namedWindow('Median Filtered Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Median Filtered Image', 600, 700)

cv2.imshow('Original Image', gray_image)
cv2.imshow('Median Filtered Image', median_filtered_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Task3
"""
I choose to use the USM.
The preference for USM over Laplacian sharpening often arises due to its simplicity, 
controllable sharpening effect, and ability to mitigate noise and unwanted artifacts effectively. 
USM's widespread adoption and ease of integration further contribute to its popularity in practical image enhancement tasks.
"""
import cv2


cv2.namedWindow('Input', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Input', 900, 600)
cv2.namedWindow('Output', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Output', 900, 600)

image = cv2.imread('./Task3.jpg', cv2.IMREAD_GRAYSCALE)

blurred_image = cv2.GaussianBlur(image, (9, 9), 10.0)

high_pass = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)

cv2.imshow('Input', image)

cv2.imshow('Output', high_pass)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
