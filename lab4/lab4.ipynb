```python
import cv2
import numpy as np
from skimage import measure, morphology, segmentation
import matplotlib.pyplot as plt

# Load image
image_path = './Fruits.png'
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply intensity thresholding
_, binary = cv2.threshold(gray, 227, 255, cv2.THRESH_BINARY_INV)

# Apply binary morphological operations
# Define the structuring element
selem = morphology.disk(2)

# Opening (erosion followed by dilation)
opened = morphology.opening(binary, selem)

# Closing (dilation followed by erosion)
closed = morphology.closing(opened, selem)

# Convert to boolean array before filling small holes
closed_bool = closed.astype(bool)

# Fill small holes inside objects
filled = morphology.remove_small_holes(closed_bool, area_threshold=500)

# Remove objects touching the image border
boundary_cleared = segmentation.clear_border(filled)

# Remove small objects to ensure final result
size_threshold = 5000  # Adjust as needed
final_cleaned = morphology.remove_small_objects(boundary_cleared, min_size=size_threshold)

# Label objects
labeled_image = measure.label(final_cleaned)

# Count the number of objects
num_objects = labeled_image.max()

# Display the result
print(f"Number of objects: {num_objects}")

# Save the result
result_path = '/mnt/data/segmented_result.png'
cv2.imwrite(result_path, (final_cleaned * 255).astype(np.uint8))
print(f"Segmented result saved to {result_path}")

# Visualize the segmented result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Segmented Result')
plt.imshow(final_cleaned, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from skimage import measure, morphology, segmentation
import matplotlib.pyplot as plt

# Load image
image_path = './Leaves.png'
image = cv2.imread(image_path)

# Convert to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range for light blue color
lower_blue = np.array([90, 50, 50])
upper_blue = np.array([100, 255, 255])

# Color filtering to keep only light blue areas
mask = cv2.inRange(hsv, lower_blue, upper_blue)
result = cv2.bitwise_and(image, image, mask=mask)

# Convert to grayscale
gray = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)

# Binarize the grayscale image
_, binary = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)

# Apply binary morphological operations
# Define the structuring element
selem = morphology.disk(1)

# Opening (erosion followed by dilation)
opened = morphology.opening(binary, selem)

# Closing (dilation followed by erosion)
closed = morphology.closing(opened, selem)

# Convert to boolean array before filling small holes
closed_bool = closed.astype(bool)

# Fill small holes inside objects
filled = morphology.remove_small_holes(closed_bool, area_threshold=500)

# Remove objects touching the image border
boundary_cleared = segmentation.clear_border(filled)

# Remove small objects to ensure final result
size_threshold = 500  # Adjust as needed
final_cleaned = morphology.remove_small_objects(boundary_cleared, min_size=size_threshold)

# Label objects
labeled_image = measure.label(final_cleaned)

# Count the number of objects
num_objects = labeled_image.max()

# Display the result
print(f"Number of objects: {num_objects}")

# Save the result
result_path = '/mnt/data/segmented_result.png'
cv2.imwrite(result_path, (final_cleaned * 255).astype(np.uint8))
print(f"Segmented result saved to {result_path}")

# Visualize the segmented result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Segmented Result')
plt.imshow(final_cleaned, cmap='gray')
plt.axis('off')

plt.show()
```
